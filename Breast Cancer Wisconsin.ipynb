```python
drop_list1 = ['perimeter_mean','radius_mean','compactness_mean','concave points_mean','radius_se','perimeter_se','radius_worst','perimeter_worst','compactness_worst','concave points_worst','compactness_se','concave points_se','texture_worst','area_worst']
x1 = x.drop(drop_list1,axis = 1 )        # do not modify x, we will use it later 
x1.head()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(x1.corr(),annot=True,fmt='.1f')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.metrics import confusion_matrix,accuracy_score as score
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
```

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y,test_size=0.2,random_state=42)
```

# Random Forest Classifier

```python
rfc=RandomForestClassifier(n_estimators=100,random_state=42)
rfc.fit(x_train,y_train)
yhat=rfc.predict(x_test)
print('the score is= ',score(y_test,yhat)*100)
sns.heatmap(confusion_matrix(y_test,yhat),annot=True,fmt='d')
```

```python
features=SelectKBest(chi2,k=7).fit(x_train,y_train)
```

```python
x_train1=features.transform(x_train)
x_test1=features.transform(x_test)

rfc1=RandomForestClassifier(random_state=42)
rfc.fit(x_train1,y_train)
yhat1=rfc.predict(x_test1)
print('the score is= ',score(y_test,yhat1)*100)
sns.heatmap(confusion_matrix(y_test,yhat1),annot=True,fmt='d')
```

```python
rfe=RFE(estimator=RandomForestClassifier(),n_features_to_select=11,step=1)
rfe.fit(x_train,y_train)
yhat2=rfe.predict(x_test)
print(score(yhat2,y_test)*100)
sns.heatmap(confusion_matrix(y_test,yhat2),annot=True,fmt='d')
```

# Logistic Regression

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
yhat_lr=lr.predict(x_test)
print(score(yhat_lr,y_test)*100)
sns.heatmap(confusion_matrix(y_test,yhat_lr),annot=True,fmt='d')
```

# DecisionTree Classifier

```python
dt=DecisionTreeClassifier(criterion='entropy')
dt.fit(x_train,y_train)
yhat_dt=lr.predict(x_test)
print(score(yhat_dt,y_test)*100)
sns.heatmap(confusion_matrix(y_test,yhat_dt),annot=True,fmt='d')
```

# Deep learning model

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from keras.optimizers import Adam
```

```python
y_train_s=LabelEncoder().fit_transform(y_train)
y_test_s=LabelEncoder().fit_transform(y_test)

ncols=x_train.shape[1]
model=Sequential()
model.add(Dense(ncols,input_shape=(ncols,),activation='sigmoid'))
model.add(Dense(12,activation='sigmoid'))
model.add(Dense(1,activation='sigmoid'))
optimizer = Adam(learning_rate=0.0001)
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train,y_train_s,epochs=2000,validation_data=(x_test,y_test_s))
```

```python
yhat_d=model.predict(x_test)
yhat_d=(yhat_d>=0.5).astype(int)
```
